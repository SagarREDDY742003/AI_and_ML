# Count words and lines in a file

```python
def count_lines(file_path):
    try:
        with open(file_path,'r') as file:
            line_count = sum(1 for line in file)
            print("Number of lines in the file:", line_count)
    except FileNotFoundError:
        print("File not found.")

def count_words(file_path):
    try:
        with open(file_path,'r') as file:
            lines = file.readlines()
            word_count = sum(len(line.split()) for line in lines)
            print("Number of words in the file:", word_count)
    except FileNotFoundError:
        print("File not found.")

file_path = './example.txt'
count_lines(file_path)
count_words(file_path)
```

# Write and Read a list of items

```python
def write_items_to_file(filename, items):
    with open(filename,'w') as file:
        for item in items:
            file.write(item + "\n")
            
def read_items_from_file(filename):
    try:
        with open(filename,'r') as file:
            items = file.readlines()
            print("Items in file are:")
            for item in items:
                print(item.strip())
    except FileNotFoundError:
        print("File not found")
        
write_items_to_file('./example.txt',["Hello World","How are you","hello Every One"])
read_items_from_file('./example.txt')
```

# write program to copy the contents of one file to another

```python
def copy_file(filename,filecopyname):
    with open(filename,'r') as file:
        content = file.read()
        with open(filecopyname,'w') as filecopy:
            filecopy.write(content)

copy_file('./example.txt','./copy.txt')
```

# Create a program that counts the number of occurances of a specific word in a text file

```python
def count_occurance_of_word_in_file(filename,word):
    with open(filename,'r') as file:
        count = file.read().lower().split().count(word.lower())
        print(count)
        
count_occurance_of_word_in_file('./example.txt','Hello')            
```

# write a program to log messages with timestamps into a file

```python
import logging

# Configure logging
logging.basicConfig(
    filename="app.log",                # Log file name
    level=logging.INFO,                 # Minimum log level
    format="%(asctime)s - %(levelname)s - %(message)s",  # Log format with timestamp
    datefmt="%Y-%m-%d %H:%M:%S"         # Timestamp format
)

# Example log messages
logging.info("Application started")
logging.warning("This is a warning message")
logging.error("An error occurred")
logging.info("Application finished")
logging.shutdown()
```
