# Create a Text Cleaner

```python
import re

def clean_text(text):
    
    # remove punctuation
    text = re.sub(r"[^\w\s]","",text)
    # remove extra spaces
    text = " ".join(text.split())
    # convert to lower case
    return text.lower()

input = "   Hello, world.!!! welcome         to python.......  "
cleaned_text = clean_text(input) 
print(cleaned_text)
    
```

# check if string is palindrome

```python
def is_palindrome(text):
    text = "".join(char.lower() for char in text if char.isalnum())
    return text == text[::-1]

text="A man, A Plan, a canal, Panama"
if(is_palindrome(text)):
    print("Palindrome")
else:
    print("Not Palindrome")
```

# Write a program to count num of vowels in a string

```python
def count_vowels(text):
    vowels = set("aeiou")
    count = 0
    for char in text.lower():
        if char in vowels:
            count+=1
    return count

text = "sagar"
print(count_vowels(text))
```

# Create a program to find and replace all email addresses in a text using regex.

```python
import re

def replace_emails(text, replacement):
    email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'
    return re.sub(email_pattern, replacement,text)


text = "Hello John, please contact us at support@example.com for help. You can also reach out to admin123@my-domain.org or sales@company.co.in."
replaced_text = replace_emails(text,"[Email]")
print(replaced_text)
```

# write a program to reverse the words in a sentence(not letters)

```python
def reverse_words(text):
    return " ".join( word[::-1] for word in text.strip().split())

print(reverse_words("hello world"))
```
