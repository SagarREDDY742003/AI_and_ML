# Word Frequency Counter

```python
string = "Hello world hello all Hello everyone"
words = string.split()
print(words)

word_count = {}

for word in words:
    word = word.lower()
    if word in word_count:
        word_count[word]+=1
    else:
        word_count[word]=1
        
print(word_count)
```

# write a program to reverse a list and remove duplicates using a set. 

```python
def reverse_a_list(list):
    return list[::-1]

def remove_duplicates(list):
    seen = set()
    unique_list = []
    
    for i in list:
        if i not in seen:
            seen.add(i)
            unique_list.append(i)
            
    return unique_list

list = [1,2,4,5,7,8,5,9,8]

reversed_list = reverse_a_list(list)
print(reversed_list)

print(remove_duplicates(reversed_list))
```

# Create a program that stores student grades in a dictionary and calculate the average grade.

```python
def calculate_average(student_grades):
    if not student_grades:
        return 0
    return sum(student_grades.values())/len(student_grades)

def get_grade_input():
    while True:
        try:
            grade = float(input("Enter Grade between 0 to 100"))
            if 0<=grade<=100 :
                return grade
            else:
                print("Grade must be between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
            

def main():
    student_grades = {}
    print("=== Student Grades Program ===")
    while True:
        name = input("Enter student name (or 'done' to finish): ").strip()
        if name.lower() == 'done':
            break
        if not name:
            print("Name cannot be empty.")
            continue
        
        grade = get_grade_input()
        student_grades[name] = grade
        
    if student_grades:
        print(student_grades)
        avg = calculate_average(student_grades)
        print("Average of Grades: ", avg)
    else:
        print("No student data entered.")
                  
if __name__ == "__main__":   
    main()
```
